```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

1. Data Cleaning


```python
# URL dataset
url = "https://raw.githubusercontent.com/Sandalgoreng/DATAMINING/refs/heads/main/customer_analysis.csv"

# Membaca dataset
data = pd.read_csv(url)
data
```

```python
# Membaca dataset dengan mencoba delimiter tab
data = pd.read_csv(url, delimiter='\t')
```

```python
# Mengatasi nilai yang hilang
data = data.fillna(data.mean(numeric_only=True))
data = data.fillna(data.mode().iloc[0])
```

```python
# Menghapus data duplikat
data = data.drop_duplicates()
```

```python
# Informasi awal dataset
print("Informasi dataset setelah cleaning:")
print(data.info())
```

2. DATA PREPERATION

```python
# Memilih fitur yang relevan
selected_columns = [
    "Year_Birth", "Education", "Marital_Status", "Income", "Kidhome", "Teenhome",
    "Recency", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts",
    "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", "NumWebPurchases",
    "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Response"
]
data = data[selected_columns]
```

```python
# Encoding variabel kategorikal dengan aman menggunakan .loc
label_enc = LabelEncoder()
categorical_cols = ["Education", "Marital_Status"]

for col in categorical_cols:
    data.loc[:, col] = label_enc.fit_transform(data[col])
```

3. Data Transformation

```python
# Normalisasi/Standarisasi kolom numerik
numerical_cols = [
    "Year_Birth", "Income", "Kidhome", "Teenhome", "Recency", "MntWines",
    "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts",
    "MntGoldProds", "NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases",
    "NumStorePurchases", "NumWebVisitsMonth"
]
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

```python
# Membagi dataset menjadi fitur (X) dan target (y)
X = data.drop("Response", axis=1)
y = data["Response"]
```

5. Clustering

```python
# Menggunakan KMeans untuk clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
# Visualisasi hasil clustering menggunakan PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)
data['PCA1'] = pca_result[:, 0]
data['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Cluster', data=data, palette='viridis', s=50
)
plt.title("Hasil Clustering dengan KMeans")
plt.show()
```

```python
# Menampilkan ringkasan hasil clustering
print("Hasil clustering (contoh 10 data):")
print(data[['Cluster', 'PCA1', 'PCA2']].head(10))
```

```python
# Standarisasi data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Hierarchical Clustering
# Menggunakan metode linkage 'ward'
linkage_matrix = linkage(data_scaled, method='ward')
```

```python
# Visualisasi Dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title("Dendrogram Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Euclidean Distance")
plt.show()

```

```python
# Menentukan jumlah cluster dari dendrogram
# Misal kita tentukan cutoff distance 6
clusters = fcluster(linkage_matrix, t=6, criterion='distance')
data['Cluster'] = clusters

```

```python
# Visualisasi hasil clustering dengan Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_scaled[:, 0], y=data_scaled[:, 1], hue=data['Cluster'], palette='viridis', s=50)
plt.title("Hierarchical Clustering Visualization")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.legend(title="Cluster")
plt.show()

```
